```python
def tic_tac_toe(board):
    
    evaluation = 0
    winner = ""

    if board == "board1":
        board = [
        ['X','X','O'],
        ['O','X','O'],
        ['O','','X'],
        ]
    elif board == "board2":
        board = [
        ['X','X','O'],
        ['O','X','O'],
        ['','O','X'],
        ]
    elif board == "board3":
        board = [
        ['O','X','O'],
        ['','O','X'],
        ['X','X','O'],
        ]
    elif board == "board4":
        board = [
        ['X','X','X'],
        ['O','X','O'],
        ['O','','O'],
        ]
    elif board == "board5":
        board = [
        ['X','X','O'],
        ['O','X','O'],
        ['X','','O'],
        ]
    elif board == "board6":
        board = [
        ['X','X','O'],
        ['O','X','O'],
        ['X','',''],
        ]
    elif board == "board7":
        board = [
        ['X','X','O',''],
        ['O','X','O','O'],
        ['X','','','O'],
        ['O','X','','']
        ]
    
    diagonal1 = [board[i][i] for i,v in enumerate(board)]
    diagonal2 = [board[2-i][i] for i,v in enumerate(board)]
    diagonal1_check = set()
    diagonal2_check = set()
    [diagonal1_check.add(s) for s in diagonal1]
    [diagonal2_check.add(s) for s in diagonal2]
    
    if len(diagonal1_check) == 1:
        winner = diagonal1_check
        evaluation += 1
    elif len(diagonal2_check) == 1:
        winner = diagonal2_check
        evaluation += 1
    
    horizontal = [x for x in board]
    vertical = [x for x in zip(*board)]
    t = 0
    
    for i in range(len(board)):
        h = horizontal[t]
        v = vertical[t]
        horizontal_check = set()
        vertical_check = set()
        [horizontal_check.add(s) for s in h]
        [vertical_check.add(s) for s in v]
        t += 1
        if len(horizontal_check) == 1 :
            winner = horizontal_check
            evaluation += 1
        elif len(vertical_check) == 1:
            winner = vertical_check
            evaluation += 1
        else:
            evaluation += 0
            
    if winner == {"X"} and evaluation >= 1:
        return "X"
    elif winner == {"O"} and evaluation >= 1:
        return "O"
    else:
        return "NO WINNER"

tic_tac_toe("board3")
```
